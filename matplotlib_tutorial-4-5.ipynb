## Stack Plots

```python
import pandas as pd
import numpy as np

from matplotlib import pyplot as plt
```

```python
plt.figure(figsize=(7, 4))

plt.style.use(['default', 'tableau-colorblind10'])
```

```python

minutes = [1, 2, 3, 4, 5, 6, 7, 8, 9]

player1 = [8, 6, 5, 5, 4, 2, 1, 1, 0]
player2 = [0, 1, 2, 2, 2, 4, 4, 4, 4]
player3 = [0, 1, 1, 1, 2, 2, 3, 3, 4]

labels = ['player1', 'player2', 'player3']
#colors = ['#6d904f', '#fc4f30', '#008fd5']

plt.stackplot(minutes, player1, player2, player3, labels=labels)

plt.xlabel("Minutes")
plt.ylabel("Points")

plt.legend(loc=(0.07, 0.05))

plt.title("Plot of Players 1-3 \n Points Scored from 1-9mins")
plt.tight_layout()
plt.show()
```

## Filling Area on Line

```python
data = pd.read_csv('data.csv')
ages = data.Age
dev_salaries = data.All_Devs
py_salaries = data.Python
js_salaries = data.JavaScript
```

```python
plt.plot(ages, dev_salaries, color= '#9B59B6',
        linestyle = '--', label = 'All Devs')

plt.plot(ages, py_salaries, label = 'Python')

overall_median = 57287
plt.fill_between(ages, py_salaries, overall_median, 
                 alpha = 0.3, interpolate=True,
                 where=(py_salaries > overall_median))

plt.fill_between(ages, py_salaries, overall_median, 
                 alpha = 0.3, interpolate=True,
                 where=(py_salaries < overall_median))

plt.legend()

plt.title("Median Salary (USD) by Age")
plt.xlabel("Ages")
plt.ylabel('Median Salary (USD)')
plt.show()
```
