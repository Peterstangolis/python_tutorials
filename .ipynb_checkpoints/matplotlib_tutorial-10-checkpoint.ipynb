## Subplots

### Import required libraries

```python
import pandas as pd

from matplotlib import pyplot as plt
```

```python
plt.figure(figsize=(7, 4))

plt.style.use(['default', 'ggplot'])
```

### Import the data set

```python
data = pd.read_csv('dev_salaries.csv')
ages = data['Age']
dev_salaries = data['All_Devs']
py_salaries = data['Python']
js_salaries = data['JavaScript']
```

### Plot the data

```python
plt.plot(ages, py_salaries, label = "Python")

plt.plot(ages, js_salaries, label = "JavaScript")

plt.plot(ages, dev_salaries, color = 'blue',
        linestyle = '--', label = 'All Devs')

plt.legend()

plt.title('Median Salary (USD) by Age')
plt.xlabel('Ages')
plt.ylabel('Median Salary (USD)')

plt.tight_layout()
plt.show()
```

## Using the Object Oriented Approach when working with plots:

* works better when you want data in multiple plots
* add more axes to do this

```python
# instantiate a figure and axis

# defaults to a 1x1 if we dont specify # of plots
fig, (ax1, ax2) = plt.subplots(nrows=2, ncols=1, sharex=True)

ax2.plot(ages, py_salaries, label = "Python")

ax2.plot(ages, js_salaries, label = "JavaScript")

ax1.plot(ages, dev_salaries,
        linestyle = '--', label = 'All Devs')

ax1.legend(facecolor='white')
ax1.set_title('Median Salary (USD) by Age')
#ax1.set_xlabel('Ages')
ax1.set_ylabel('Median Salary (USD)')

ax2.legend(facecolor='white', framealpha = 0.7)
#ax2.set_title('Median Salary (USD) by Age')
ax2.set_xlabel('Ages')
ax2.set_ylabel('Median Salary (USD)')

plt.tight_layout()
plt.show()
```

```python
fig, ax = plt.subplots(nrows=2, ncols=1)

print(ax)
```

## To place the plots of 2 separate figures:

```python
# defaults to a 1x1 if we dont specify # of plots
fig1, ax1 = plt.subplots()
fig2, ax2 = plt.subplots()

ax2.plot(ages, py_salaries, label = "Python")

ax2.plot(ages, js_salaries, label = "JavaScript")

ax1.plot(ages, dev_salaries,
        linestyle = '--', label = 'All Devs')

ax1.legend(facecolor='white')
ax1.set_title('Median Salary (USD) by Age')
ax1.set_xlabel('Ages')
ax1.set_ylabel('Median Salary (USD)')

ax2.legend(facecolor='white', framealpha = 0.7)
ax2.set_title('Median Salary (USD) by Age')
ax2.set_xlabel('Ages')
ax2.set_ylabel('Median Salary (USD)')

plt.tight_layout()
plt.show()
```
