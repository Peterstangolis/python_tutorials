## Plotting Time Series Data

```python
import pandas as pd

from datetime import datetime, timedelta
from matplotlib import pyplot as plt
from matplotlib import dates as mpl_dates
```

```python
plt.figure(figsize=(7, 4))

plt.style.use(['default', 'dark_background'])
```

```python
print(plt.style.available)
```

```python
data = pd.read_csv('bitcoin.csv')

data['Date'] = pd.to_datetime(data['Date'])
data.sort_values('Date', inplace=True)

price_date = data['Date']
price_close = data['Close']

plt.plot_date(price_date, price_close, 
              linestyle='solid',
              color="#229954")

# Get Current Figure (dates are rotated and aligned)
plt.gcf().autofmt_xdate()

#date_format = mpl_dates.DateFormatter('%b, %d, %Y')
# Get Current Axis
#plt.gca().xaxis.set_major_formatter(date_format)


plt.title('Bitcoin Prices', fontsize=16, color='y')
#plt.xlabel('Date')
plt.ylabel('Closing Price')

plt.tight_layout()
#plt.grid(True)

plt.show()
```

## Plotting Live Data in Real-Time

```python
from itertools import count
import random

from matplotlib.animation import FuncAnimation
%matplotlib qt
```

```python
plt.style.use(['default', 'seaborn-paper'])
```

```python
x_vals = []
y_vals = []

index = count()

def animate(i):
    x_vals.append(next(index))
    y_vals.append(random.randint(0, 5))
    
    plt.cla()
    plt.plot(x_vals, y_vals)

anim = FuncAnimation(plt.gcf(), animate, interval=1000)
    

plt.tight_layout()
plt.show()
```
